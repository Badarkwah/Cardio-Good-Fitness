-   Education: - About 50% of the customers have 16 years of education
    out of the maximum number of 21 years. Looking the spread of
    educutions, I believe it's safe to say that the customers for this
    fitness products have at least high school education or 12 years.
-   Income- The average customer for these products makes a little over
    53,719USD in income, with most income being 104,581USD
-   Fitness - The average person is kind of fit with a score of 3 on a
    1-5 gradient. 1 being unfit and 5 being very fit. 75% of the
    customers recorded 4 with 25% of the customers recorded 3.
-   Usage: - 50% of the customers will love to use the treadmill 3 times
    a week while 75% will love to use it 4 times a week.
